```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
```

```python

place_name = "Munich"
# Get place boundary related to the place name as a geodataframe
area = ox.geocode_to_gdf(place_name)
```

```python
area.plot()
```

```python
# List key-value pairs for tags
tags = {
    'building': True, 
}   

buildings = ox.geometries_from_place(place_name, tags,)
```

```python
roads = tags = {
    "highway": True
}   

roads = ox.geometries_from_place(place_name, tags,)
```

```python
tags = {
    'club': 'sport',
    'amenity': 'language_school',
    'amenity': 'library',
    'public_transport':'station',
    #"highway": True
}   

extras = ox.geometries_from_place(place_name, tags,)
```

```python
extras.to_crs(area.crs)
area_base = area.plot(color='white', edgecolor='black')
roads_base = roads.plot(ax=area_base, color="grey")

extras.plot(ax=roads_base, figsize=(10,100))
```
